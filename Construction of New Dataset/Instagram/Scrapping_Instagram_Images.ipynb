```python
bot = instaloader.Instaloader(download_pictures=True, download_videos=False, download_video_thumbnails=False, 
                              download_geotags=False, download_comments=False, save_metadata=True, compress_json=False, 
                              post_metadata_txt_pattern=None, storyitem_metadata_txt_pattern=None, max_connection_attempts=3,
                              request_timeout=None, rate_controller=None, resume_prefix='iterator', check_resume_bbd=True)
profile = instaloader.Profile.from_username(bot.context, 'Edtsousa')
```

```python
posts = profile.get_posts()
```

```python
# Downloading all the images and storing them in one folder
for index, post in enumerate(posts,1):
    bot.download_post(post, target = "%s"%(index-1))
```

```python
all_cap_files = glob.glob('./**/*.txt')
```

```python
# Reading all the .txt file to extract captions and storing them in a list
txt = []
for file in all_cap_files:
    f= open(file,'r')
    t = f.read()
    txt.append(t)
```

```python
len(txt)
```

```python
# Reading all the .json files and storing their values in a list
all_js_files = glob.glob('./**/*.json')
js= []
for file in all_js_files:
    with open(file) as f:
        js.append(json.load(f))
```

```python
# Saving the data in required format
coco = []
imgs = []
annos = []
for (index,item) in enumerate(txt):
    if(len(imgs)>199):
        break
    try:
        
        dict_ = {
                "id":index,
                "width":js[index]['node']['dimensions']['width'],
                "height":js[index]['node']['dimensions']['height'],
                "file_name":str(index)+".jpeg"
            },
        annotation = {
                "id":index,
                "image_id":index,
                "caption":item
            }
        imgs.append(dict_)
        annos.append(annotation)

    except: 
        continue
nest = {
        "images":imgs,
        "annotations":annos
    }
```

```python
nest
```

```python
# Saving the file
with open('./final_captions_insta.json', 'w') as json_file:
    json.dump(nest, json_file)
```

```python
# Saving file with just removing hastag
with open('/captions_insta2020.json') as f:
  data = json.load(f)
for i in range(len(data['annotations'])):
    val = data['annotations'][i]['caption']
    data['annotations'][i]['caption'] = val.replace('#','')
# Saving the file
with open('/captions_insta2020_no.json', 'w') as json_file:
    json.dump(data, json_file)
```

```python
# Creating and saving file with removing words starting with hastag
import re,json
with open('/captions_insta2020.json') as f:
  data = json.load(f)
for i in range(len(data['annotations'])):
    val = data['annotations'][i]['caption']
    data['annotations'][i]['caption'] = re.sub('#.*','',val)
# Saving the file
with open('/captions_insta2020_yes.json', 'w') as json_file:
    json.dump(data, json_file)
```
